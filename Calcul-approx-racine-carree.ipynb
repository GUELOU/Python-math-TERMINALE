Ce programme calcule une valeur approchée de la racine carré d'une nombre entier positif A saisi par l'utilisateur.
Par défaut, le programme réalise 100 itérations avant d'afficher le résultat.
On utilise ici la suite définie par Un+1=0.5(Un+A/Un) et U0=A qui converge vers sqrt(A)

```python
from math import *

n=0
A=int(input("Saisir la valeur A="))
u=A
print ("Liste des 100 premiers termes de la suite")
while n<100:
    u=0.5*(u+A/u)#définition de la suite par récurrence
    n=n+1
    print(u)
print("une valeur approchée de sqrt(",A,")est :",u)
```
