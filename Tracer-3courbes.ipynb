Représentation Graphique d'une fonction mathématique

Pour tracer des courbes la fonction %pylab inline importe les bibliothèques Numpy et Matplotlib.
Il devient possible de dessiner des graphiques intégrés au notebook.

```python
%pylab inline

x =linspace(0, 2, 100)

y=x
z=x*x
t=x*x*x

grid()

plot(x,y, label='linear')
plot(x,z, label='quadratic')
plot(x,t, label='cubic')

xlabel('x label')
ylabel('y label')
title("Simple Plot")

legend()

savefig('graph3fonctions.png') # écriture du graphique dans un fichier image

```
