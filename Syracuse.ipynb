Ce programme sert à tester la conjecture de Collatz:  
(aussi appelée conjecture de Syracuse, ou "hailstone sequence" en anglais)  

Soit N un nombre entier   
Si N est pair alors le terme suivant est N/2  
Si N est impair alors le terme suivant est 3N+1  

La Conjecture est la suivante :  
"Quel que soit le nombre entier choisi au départ, la suite finit toujours par la séquence 4,2,1..."  

A ce jour, elle n'a pas été démontrée...


```python
import matplotlib as mpl
mpl.use('Agg')

import matplotlib.pyplot as plt
fig = plt.figure()
plt.grid()

print ("Hailstone Sequence")
n=input ("Enter the number you want to try:")
n=int(n)

c=0
m=0
y=n
while n>1:
  c=c+1
  plt.plot(c,y,'bo')

  if n%2==0:
   n=n/2
   print(n)
   y=n
  else:
    n=3*n+1
    print (n)
    y=n
  if y>m:
    m=y
    
print("Number of steps to reach 1: ",c)
print("Maximum reached number ",m)
fig.savefig('graph-syracuse.png') # ecriture du graphique dans un fichier image    
```
